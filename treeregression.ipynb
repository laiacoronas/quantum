```python
import pandas as pd
data = pd.read_csv("roboBohr.csv")
data.head(5)
```

```python
cols = [col for col in data.columns if 'Unnamed' not in col and 'pubchem' not in col]
data = data[cols]
data.head(5)
```

```python
data.shape
```

```python
#machine learning
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.metrics import accuracy_score
from xgboost import XGBRegressor

X = data.drop('Eat', axis=1)
y = data['Eat']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)

param_grid = {
    'n_estimators': [400, 600],
    'learning_rate': [0.01, 0.1],
    'reg_alpha': [0],
    'max_depth': [6],
    'subsample': [0.2],
    'min_child_weight': [10]
}

xgb_model = XGBRegressor()

grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_

best_model = XGBRegressor(**best_params)
best_model.fit(X_train, y_train)

predictions = best_model.predict(X_val)
y_val_array = y_val.values.astype(np.float32)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_val, predictions)
print("Mean Squared Error:", mse)
```

```python
import requests
from io import BytesIO
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import PandasTools

def get_coulomb_matrix(pubchem_id):
    # Descargar el archivo SDF desde PubChem
    response = requests.get(f"https://pubchem.ncbi.nlm.nih.gov/rest/pug/compound/cid/{pubchem_id}/SDF")
    response.raise_for_status()  # Lanza una excepciÃ³n si la solicitud falla

    # Leer el contenido del archivo SDF
    sdf_content = response.content

    # Cargar el archivo SDF en un objeto Mol
    suppl = Chem.SDMolSupplier()
    suppl.SetData(BytesIO(sdf_content))
    mol = suppl[0]

    # AÃ±adir hidrÃ³genos y optimizar la geometrÃ­a
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol)
    AllChem.UFFOptimizeMolecule(mol)

    # Calcular la matriz de Coulomb
    cm = AllChem.GetMolecularCoulombMatrix(mol)

    return cm

pubchem_id = "750"  
coulomb_matrix = get_coulomb_matrix(pubchem_id)
print("Matriz de Coulomb:")
print(coulomb_matrix)
```
