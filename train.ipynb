Basic Molecular Properties:

cid: Unique identifier for the molecule in the PubChem database.

mw: Molecular weight of the molecule.

mf: Molecular formula of the molecule (e.g., C6H12O6 for glucose).

polararea: Polar surface area of the molecule, a measure of its polarity.

complexity: Measure of the molecule's structural complexity.

Pharmacophore Features:

xlogp: Predicted logarithm of the octanol-water partition coefficient, a measure of a molecule's lipophilicity (affinity for fat).

heavycnt: Number of heavy atoms (excluding hydrogen) in the molecule.

hbonddonor: Number of hydrogen bond donor sites in the molecule.

hbondacc: Number of hydrogen bond acceptor sites in the molecule.

rotbonds: Number of rotatable bonds in the molecule, important for flexibility.

Mass and Charge:

exactmass: Exact mass of the molecule (considering isotope masses).

monoisotopicmass: Mass of the molecule considering the most abundant isotope for each element.

charge: Formal charge of the molecule.

Atom and Bond Stereochemistry:

covalentunitcnt: Number of covalent unit bonds in the molecule.

isotopeatomcnt: Number of atoms with isotopes specified in the molecule.

totalatomstereocnt: Total number of stereocenters (atoms with defined chirality) in the molecule.

definedatomstereocnt: Number of stereocenters with a defined chirality.

undefinedatomstereocnt: Number of stereocenters with undefined chirality.

totalbondstereocnt: Total number of double bonds with specified stereochemistry (E/Z).

definedbondstereocnt: Number of double bonds with defined stereochemistry.

undefinedbondstereocnt: Number of double bonds with undefined stereochemistry.

```python
import re

def num_qubits(formula):
  num_qubits = 0
  for atom, count in re.findall(r'([A-Z][a-z]?)([0-9]*)', formula):
  
    if atom == 'C':
      multiplier = 6
    elif atom == 'O':
      multiplier = 8
    elif atom == 'H':
      multiplier = 1
    elif atom == 'N':
      multiplier = 7  
    elif atom == 'S':
      multiplier = 16

    count = int(count) if count else 1
    num_qubits += count * multiplier
  return num_qubits
```

```python
pub['Num Qubits'] = pub['mf'].apply(num_qubits)
```

```python

def ordenar_formula(formula):
    # Definir un diccionario para contener el recuento de cada elemento
    elementos = {'C': 0, 'H': 0, 'N': 0, 'O': 0, 'S': 0}
    
    # Utilizar expresiones regulares para encontrar elementos y sus recuentos
    matches = re.findall(r'([A-Z][a-z]*)(\d*)', formula)
    
    # Iterar sobre los elementos encontrados y actualizar el diccionario de recuentos
    for elemento, cantidad in matches:
        cantidad = int(cantidad) if cantidad else 1
        elementos[elemento] += cantidad
    
    # Crear la fórmula ordenada
    formula_ordenada = ''
    for elemento, cantidad in elementos.items():
        if cantidad > 0:
            formula_ordenada += elemento
            if cantidad > 1:
                formula_ordenada += str(cantidad)
    
    return formula_ordenada
```

```python
aa['Combined'] = aa['MOLECULE'].apply(ordenar_formula)
```
