```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np


# Extract the relevant columns
mw = data['NumQubits'].values.reshape(-1, 1)
Eat = data['GSE'].values

# Perform linear regression
model = LinearRegression()
model.fit(mw, Eat)
slope = model.coef_[0]
intercept = model.intercept_

# Plot Eat as a function of mw with regression line
plt.figure(figsize=(10, 6))
plt.scatter(mw, Eat, alpha=0.7, label='Data points')
plt.plot(mw, model.predict(mw), color='red', label='Regression line')
plt.title('Ground state energy as function of the number of qubits')
plt.xlabel('Number of qubits')
plt.ylabel('Ground State Energy [Ha]')
plt.legend()
plt.show()

# Print slope and intercept
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')


slope, intercept
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np


# Extract the relevant columns
mw = data['NumQubits'].values.reshape(-1, 1)
Eat = data['GSE'].values

# Perform linear regression
model = LinearRegression()
model.fit(mw, Eat)
slope = model.coef_[0]
intercept = model.intercept_

# Predict using the linear model
Eat_pred = model.predict(mw)

# Calculate the relative error percentage
relative_error = np.abs((Eat_pred - Eat) / Eat) * 100

# Add predictions and relative error to the dataframe
data['Eat_pred'] = Eat_pred
data['Relative_Error_%'] = relative_error

# Print the first few rows to see the results
print(data.head())


# Print the slope and intercept
print(f'Slope: {slope}')
print(f'Intercept: {intercept}')
```
