```python
import re
def invert_partial_id_order(id_string):
  """Inverts the order of the first part of an ID string while preserving the rest."""
  pattern = r"([A-Za-z][0-9]*)(.*)"  # Regular expression pattern to capture the first part
  match = re.match(pattern, id_string)
  if match:
    return match.group(2) + match.group(1)  # Invert and join the captured parts
  else:
    return id_string  # Return the original string if it doesn't match the pattern

# Apply the function to both DataFrames

qm7['mf'] = qm7['mf'].apply(invert_partial_id_order)

training['mf'] = training['mf'].apply(invert_id_order)
```
