```python
cols = [' cid','cmpdname', 'mw', 'polararea',
       'complexity', 'xlogp', 'heavycnt', 'hbonddonor', 'hbondacc', 'rotbonds',
       'exactmass', 'monoisotopicmass', 'charge', 'covalentunitcnt',
       'isotopeatomcnt', 'totalatomstereocnt', 'definedatomstereocnt',
       'undefinedatomstereocnt', 'totalbondstereocnt', 'definedbondstereocnt',
       'undefinedbondstereocnt']
pubchem = pubchem[cols]
pubchem
```

```python
aa = aa.rename(columns={'IDS':' cid'})
df_aa = pd.merge(aa, pubchem, on=' cid', how='left')
df_aa
```

```python
import re
def num_qubits(formula):
  num_qubits = 0
  for atom, count in re.findall(r'([A-Z][a-z]?)([0-9]*)', formula):
    if atom == 'C':
      multiplier = 6
    elif atom == 'O':
      multiplier = 8
    elif atom == 'H':
      multiplier = 1
    elif atom == 'N':
      multiplier = 7  
    elif atom == 'S':
      multiplier = 16
    count = int(count) if count else 1
    num_qubits += count * multiplier
  return num_qubits
```

```python
df_aa['qubits'] = df_aa['FORMULA'].apply(num_qubits)
df_aa
```

```python
import numpy as np
print(np.max(df_aa['qubits']))
print(np.mean(df_aa['qubits']))
print(np.min(df_aa['qubits']))
```

```python
def contar_atomos(molecula, elemento):
    return molecula.count(elemento)
for elemento in ['C', 'H', 'O', 'N']:
    df_aa[elemento] = df_aa['FORMULA'].apply(contar_atomos, args=(elemento,)) #one hot encoding
df_aa
```
