```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()



xtrain = train.rename(columns={'Num Qubits': 'NumQubits'})

train = train.drop(['pubchem_id','mf'], axis = 1)

x = scaler.fit_transform(train)

xscaled = pd.DataFrame(x, columns=train.columns)

ytrain = xscaled[['Eat']]

xscaled = xscaled.drop(['Eat'], axis = 1)


```

```python
train_cols = set(xscaled.columns)
test_cols = set(X.columns)

# Agregar columnas faltantes con ceros
for col in train_cols - test_cols:
    X[col] = 0

for col in test_cols - train_cols:
    xscaled[col] = 0

# Asegurarse de que las columnas estén en el mismo orden
X = X[xscaled.columns]
```

```python
xscaled
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error
from xgboost.sklearn import XGBRegressor
import lightgbm as lgb
model = XGBRegressor(
    learning_rate = 0.1, reg_lambda = 0, max_depth = 5, subsample = 0.8, colsample_bytree = 0.2, min_child_weight = 10, n_estimators = 200

)
model.fit(xscaled, ytrain)
ypredict = model.predict(X)

mape_scores = []

n_iterations = 10  # Número de iteraciones

for _ in range(n_iterations):
    # Entrenar el modelo
    model.fit(xscaled, ytrain)

    # Realizar predicciones en el conjunto de prueba
    y_pred = model.predict(X)

    # Calcular MAPE en el conjunto de prueba
    mape = mean_absolute_percentage_error(y, y_pred)
    mape_scores.append(mape)

# Calcular la media y la desviación estándar del MAPE
mean_mape = np.mean(mape_scores) * 100  # Convertir a porcentaje
std_mape = np.std(mape_scores) * 100  # Convertir a porcentaje

print(f"Mean MAPE: {mean_mape:.2f}%")
print(f"STD of MAPE: {std_mape:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from xgboost.sklearn import XGBRegressor


from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(xscaled, ytrain, test_size=0.2, random_state=42)

model = XGBRegressor(
    learning_rate=0.1, reg_lambda=0, max_depth=5, subsample=0.8,
    colsample_bytree=0.2, min_child_weight=10, n_estimators=200
)

eval_set = [(x_val, y_val)]
eval_metric = 'rmse'
model.fit(x_train, y_train, eval_metric=eval_metric, eval_set=eval_set, verbose=False)

# Extraer el RMSE del historial de evaluación
results = model.evals_result()
rmse = results['validation_0']['rmse']

# Plot del RMSE por número de iteraciones
plt.plot(rmse)
plt.xlabel('Number of iterations')
plt.ylabel('RMSE')
plt.title('RMSE training')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from xgboost.sklearn import XGBRegressor
from sklearn.model_selection import train_test_split

# Supongamos que xscaled y ytrain ya están definidos
# Divide los datos en conjuntos de entrenamiento y validación
x_train, x_val, y_train, y_val = train_test_split(xscaled, ytrain, test_size=0.2, random_state=42)

model = XGBRegressor(
    learning_rate=0.1, reg_lambda=0, max_depth=5, subsample=0.8,
    colsample_bytree=0.2, min_child_weight=10, n_estimators=500
)

# Entrenar el modelo y evaluar el RMSE en cada iteración
eval_set = [(x_val, y_val)]
eval_metric = 'rmse'
model.fit(x_train, y_train, eval_metric=eval_metric, eval_set=eval_set, verbose=False)

# Extraer el RMSE del historial de evaluación
results = model.evals_result()
rmse = results['validation_0']['rmse']

# Generar predicciones
predictions = model.predict(x_val)

# Plot del RMSE por número de iteraciones
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.plot(rmse)
plt.xlabel('Iterations')
plt.ylabel('RMSE')
plt.title('RMSE per number of iterations during the training phase')

# Plot de las predicciones
plt.subplot(1, 2, 2)
plt.scatter(range(len(y_val)), y_val, color='blue', label='Ground Truth')
plt.scatter(range(len(y_val)), predictions, color='red', alpha=0.5, label='Predictions')
plt.xlabel('Index')
plt.ylabel('Atomization Energies')
plt.title('Ground Truths vs Predictions')
plt.legend()

plt.tight_layout()
plt.show()
```
