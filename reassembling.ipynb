```python
import pandas as pd
data = pd.read_csv("C:/Users/lclai/Desktop/LDIG/DB/aa/def/fragments.csv")
data
```

```python
import numpy as np
# Suma de las energías de los fragmentos
data['Sum_Energy'] = data[['1', '2', '3', '4']].sum(axis=1)

# Promedio ponderado de las energías de los fragmentos basado en el número de qubits
weights = data[['q1', 'q2', 'q3', 'q4']]
energies = data[['1', '2', '3', '4']]
data['Weighted_Avg_Energy'] = np.average(energies, weights=weights, axis=1)

# Error relativo entre la energía calculada y la energía real (GSE)
data['Relative_Error_Sum'] = np.abs((data['Sum_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Weighted'] = np.abs((data['Weighted_Avg_Energy'] - data['GSE']) / data['GSE']) * 100


print(np.mean(data['Relative_Error_Sum']))

print(np.mean(data['Relative_Error_Weighted']))
```

```python
data['Custom_Operation'] = data['1'] + data['2'] + data['3'] +( data['q4'] * data['4'] ) / data['QUBITS']
```

```python
data['RE'] = np.abs((data['Custom_Operation'] - data['GSE']) / data['GSE']) * 100
print(np.mean(data['RE']))
data[['GSE', 'Custom_Operation']]
```

```python
from scipy.stats import gmean
data['Mean_Energy'] = data[['1', '2', '3', '4']].mean(axis=1)

data['Geometric_Mean_Energy'] = data[['1', '2', '3', '4']].apply(gmean, axis=1)

data['Median_Energy'] = data[['1', '2', '3', '4']].median(axis=1)

adjusted_weights = np.exp(data[['q1', 'q2', 'q3', 'q4']])
data['Adjusted_Weighted_Avg_Energy'] = np.average(data[['1', '2', '3', '4']], weights=adjusted_weights, axis=1)

data['Relative_Error_Mean'] = np.abs((data['Mean_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Geometric_Mean'] = np.abs((data['Geometric_Mean_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Median'] = np.abs((data['Median_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Adjusted_Weighted'] = np.abs((data['Adjusted_Weighted_Avg_Energy'] - data['GSE']) / data['GSE']) * 100
```

```python
# Error relativo entre la energía calculada y la energía real (GSE) para cada método
data['Relative_Error_Sum'] = np.abs((data['Sum_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Weighted'] = np.abs((data['Weighted_Avg_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Mean'] = np.abs((data['Mean_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Median'] = np.abs((data['Median_Energy'] - data['GSE']) / data['GSE']) * 100
data['Relative_Error_Adjusted_Weighted'] = np.abs((data['Adjusted_Weighted_Avg_Energy'] - data['GSE']) / data['GSE']) * 100

# Calcular las medias de los errores relativos para cada método de agregación
mean_relative_error_sum = data['Relative_Error_Sum'].mean()
mean_relative_error_weighted = data['Relative_Error_Weighted'].mean()
mean_relative_error_mean = data['Relative_Error_Mean'].mean()
mean_relative_error_median = data['Relative_Error_Median'].mean()
mean_relative_error_adjusted_weighted = data['Relative_Error_Adjusted_Weighted'].mean()

# Imprimir las medias de los errores relativos
print(f"Mean Relative Error (Sum): {mean_relative_error_sum:.2f}%")
print(f"Mean Relative Error (Weighted): {mean_relative_error_weighted:.2f}%")
print(f"Mean Relative Error (Arithmetic Mean): {mean_relative_error_mean:.2f}%")
print(f"Mean Relative Error (Median): {mean_relative_error_median:.2f}%")
print(f"Mean Relative Error (Adjusted Weighted): {mean_relative_error_adjusted_weighted:.2f}%")
```

```python
from scipy.optimize import minimize

def objective(params):
    a, b, c, d = params
    data['Custom_Energy'] = a * data['1'] + b * data['2'] + c * data['3'] + d * data['4']
    data['Relative_Error_Custom'] = np.abs((data['Custom_Energy'] - data['GSE']) / data['GSE']) * 100
    return data['Relative_Error_Custom'].mean()

initial_params = [1, 1, 1, 1]

result = minimize(objective, initial_params, method='BFGS')

optimal_params = result.x
print(f"Optimal parameters: a={optimal_params[0]}, b={optimal_params[1]}, c={optimal_params[2]}, d={optimal_params[3]}")

data['Custom_Energy'] = optimal_params[0] * data['1'] + optimal_params[1] * data['2'] + optimal_params[2] * data['3'] + optimal_params[3] * data['4']
data['Relative_Error_Custom'] = np.abs((data['Custom_Energy'] - data['GSE']) / data['GSE']) * 100
mean_relative_error_custom = data['Relative_Error_Custom'].mean()

print(f"Mean Relative Error (Custom): {mean_relative_error_custom:.2f}%")
```
